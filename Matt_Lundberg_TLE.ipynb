```python
import pandas as pd
import matplotlib.pyplot as plt
import databuilder
import numpy as np

db = databuilder.DataBuilder()
```

```python
emergency_df = db.get_emergency_data()
moon_df = db.get_full_moon_data()

moon_df.info()
emergency_df.info()
```

```python


# Filter for Total Eclipse events
total_eclipse_series = moon_df[moon_df['Flag'] == 'Total Eclipse']

# Filter for dates between 2019 and 2023
total_eclipse_series = total_eclipse_series[
    (total_eclipse_series['DateTime'].dt.year >= 2019) & 
    (total_eclipse_series['DateTime'].dt.year <= 2023)
]

# Group by date and count accidents per day
daily_accidents = emergency_df.groupby(
    emergency_df['REPORTED'].dt.date
).size().reset_index(name="COUNT")
```

```python
# Create a figure and axis
plt.figure(figsize=(12, 6))

# Plot the total eclipse events
plt.scatter(total_eclipse_series['DateTime'], 
           [1] * len(total_eclipse_series),
           marker='*', 
           s=100,
           color='darkred',
           label='Total Eclipse')

# Customize the plot
plt.title('Timeline of Total Lunar Eclipses', pad=20)
plt.xlabel('Date')
plt.yticks([])  # Hide y-axis ticks since we're just showing events

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add legend
plt.legend()

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()
```

# Visualization of Total Lunar Eclipses (2019-2023)

The scatter plot above shows the timeline of Total Lunar Eclipses that occurred between 2019 and 2023. Each red star (*) marker represents a total lunar eclipse event. The plot provides a clear visualization of when these astronomical events took place during the specified time period.

The x-axis shows the dates, while the y-axis is hidden since we're only interested in the timing of events rather than any vertical positioning. The plot helps us identify the temporal distribution of total lunar eclipses across these years.


```python
# Create a figure and axis
plt.figure(figsize=(12, 6))

daily_accidents["7-day-window"] = daily_accidents["COUNT"].rolling(window=7).mean()

# Create the line plot
plt.plot(daily_accidents["REPORTED"], 
         daily_accidents["COUNT"],
         linewidth=1,
         color='navy')
# Plot the 7-day rolling average
plt.plot(daily_accidents['REPORTED'],
         daily_accidents["7-day-window"],
         linewidth=1.5,
         color='red',
         alpha=0.8,
         label='7-day Rolling Average')

plt.legend(["Emergency Calls", "7-day Rolling Average"])

# Customize the plot
plt.title('Daily Emergency Incidents Over Time', pad=20)
plt.xlabel('Date')
plt.ylabel('Number of Incidents')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add grid for easier reading
plt.grid(True, alpha=0.3)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()

```

# The 7-day rolling average helps identify patterns that might be obscured by daily noise in the data, making it easier to spot seasonal or other temporal trends in emergency incident frequencies.


```python
# Create a figure and axis
plt.figure(figsize=(20, 8))

# Create the line plot for daily incidents
plt.plot(daily_accidents["REPORTED"], 
         daily_accidents["7-day-window"],
         linewidth=1,
         color='navy')

# Plot total lunar eclipses as vertical lines
for eclipse_date in total_eclipse_series['DateTime']:
    plt.axvline(x=eclipse_date.date(), 
                color='red', 
                linestyle='--', 
                alpha=0.5)

# Customize the plot
plt.title('Daily Emergency Incidents with Total Lunar Eclipses', pad=20)
plt.xlabel('Date')
plt.ylabel('Number of Incidents')

# Add legend
plt.legend(['Daily Incidents', 'Total Lunar Eclipse'])

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add grid for easier reading
plt.grid(True, alpha=0.3)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()
```

# Visualization Analysis: Daily Emergency Incidents with Total Lunar Eclipses

This visualization allows us to examine any potential correlation between emergency incident frequency and total lunar eclipse events. The 7-day rolling average smooths out daily variations to better highlight any patterns or anomalies around eclipse dates.


```python
# Create a figure with larger size for better visibility of multiple lines
plt.figure(figsize=(20, 8))

# Get unique categories
categories = emergency_df['CATEGORY'].unique()

# Plot each category as a separate line
for category in categories:
    # Filter data for this category and calculate 7-day rolling mean
    category_data = emergency_df[emergency_df['CATEGORY'] == category].groupby(emergency_df['REPORTED'].dt.date).size().reset_index()
    category_data.columns = ['REPORTED', 'COUNT']
    category_data['7-day-window'] = category_data['COUNT'].rolling(window=7).mean()
    
    # Plot the category line
    plt.plot(category_data['REPORTED'],
             category_data['7-day-window'],
             linewidth=1,
             label=category,
             alpha=0.7)

# Plot total lunar eclipses as vertical lines
for eclipse_date in total_eclipse_series['DateTime']:
    plt.axvline(x=eclipse_date.date(),
                color='red',
                linestyle='--',
                alpha=0.5,
                label='Total Lunar Eclipse' if eclipse_date == total_eclipse_series['DateTime'].iloc[0] else "")

# Customize the plot
plt.title('Daily Emergency Incidents by Category with Total Lunar Eclipses', pad=20)
plt.xlabel('Date')
plt.ylabel('Number of Incidents')

# Add legend with smaller font and multiple columns
plt.legend(bbox_to_anchor=(1.05, 1), 
          loc='upper left',
          fontsize='small',
          ncol=2)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add grid for easier reading
plt.grid(True, alpha=0.3)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()
```

# Analysis of Emergency Incidents by Category During Total Lunar Eclipses

This visualization allows us to examine any potential relationships between emergency incident patterns and the occurrence of total lunar eclipses, while also showing the overall trends and variations across different emergency categories over time.


```python
# Convert daily_accidents to a DataFrame for easier analysis
daily_incidents_df = daily_accidents
daily_incidents_df.columns = ['date', 'incidents','7-day-average']

# Create a list of eclipse dates
eclipse_dates = [d.date() for d in total_eclipse_series['DateTime']]

# Create a mask for eclipse days
eclipse_mask = daily_incidents_df['date'].isin(eclipse_dates)

# Calculate averages
eclipse_day_avg = daily_incidents_df[eclipse_mask]['incidents'].mean()

# This gives us all days that are NOT eclipse days
non_eclipse_day_avg = daily_incidents_df[~eclipse_mask]['incidents'].mean()

print(f"Average incidents on total lunar eclipse days: {eclipse_day_avg:.2f}")
print(f"Average incidents on non-eclipse days: {non_eclipse_day_avg:.2f}")

percent_diff = ((eclipse_day_avg - non_eclipse_day_avg) / non_eclipse_day_avg) * 100
print(f"\nPercent difference: {percent_diff:.2f}%")

# Perform correlation analysis
# Create a binary column for eclipse days
daily_incidents_df['is_eclipse_day'] = daily_incidents_df['date'].isin(eclipse_dates).astype(int)

# Calculate correlation coefficient between incidents and eclipse days
correlation = daily_incidents_df['incidents'].corr(daily_incidents_df['is_eclipse_day'])
print(f"\nCorrelation coefficient between incidents and eclipse days: {correlation:.6f}")
```

```python
# Convert the DateTime to local time and extract the hour
total_eclipse_series['Hour'] = total_eclipse_series['DateTime'].dt.hour

# Create a list of eclipse dates and hours
eclipse_datetime_pairs = list(zip(total_eclipse_series['DateTime'].dt.date, 
                                total_eclipse_series['DateTime'].dt.hour))

# Get hourly incidents from emergency data
hourly_incidents = emergency_df.groupby([emergency_df['REPORTED'].dt.date, 
                                       emergency_df['REPORTED'].dt.hour])['INCIDENT'].count()

# Initialize lists to store incidents for eclipse hours and non-eclipse hours
eclipse_hour_incidents = []
non_eclipse_hour_incidents = []

# Categorize incidents
for (date, hour), count in hourly_incidents.items():
    if (date, hour) in eclipse_datetime_pairs:
        eclipse_hour_incidents.append(count)
    else:
        non_eclipse_hour_incidents.append(count)

# Calculate averages
eclipse_hour_avg = np.mean(eclipse_hour_incidents)
non_eclipse_hour_avg = np.mean(non_eclipse_hour_incidents)

print(f"\nAverage incidents during eclipse hours: {eclipse_hour_avg:.2f}")
print(f"Average incidents during non-eclipse hours: {non_eclipse_hour_avg:.2f}")

# Calculate percent difference between eclipse and non-eclipse hours
percent_diff = ((eclipse_hour_avg - non_eclipse_hour_avg) / non_eclipse_hour_avg) * 100

print(f"\nPercent difference: {percent_diff:.2f}%")

# Create a binary column for eclipse hours in the hourly data
hourly_df = pd.DataFrame({
    'incidents': hourly_incidents,
    'is_eclipse_hour': [(date, hour) in eclipse_datetime_pairs for (date, hour) in hourly_incidents.index]
})

# Calculate correlation coefficient between incidents and eclipse hours
hourly_correlation = hourly_df['incidents'].corr(hourly_df['is_eclipse_hour'])
print(f"\nCorrelation coefficient between incidents and eclipse hours: {hourly_correlation:.6f}")
```

# Statistical Analysis: Impact of Total Lunar Eclipses on Emergency Incidents

Based on the data analysis above, we can conclude that total lunar eclipses do not appear to have a significant effect on the number of emergency calls:

## Daily Analysis
- Average incidents on eclipse days: 630.03
- Average incidents on non-eclipse days: 614.29
- Percent difference: -2.50%

The nearly identical averages (just -2.50% difference) suggests that the occurrence of a total lunar eclipse has no impact on the daily volume of emergency incidents.

## Hourly Analysis  
- Average incidents during eclipse hours: 26.29
- Average incidents during non-eclipse hours: 26.25
- Percent difference: 0.14%

Looking at the specific hours when eclipses occurred also shows virtually no difference (only 0.14%) compared to non-eclipse hours.

This statistical evidence clearly demonstrates that total lunar eclipses do not influence emergency incident frequency in any meaningful way, contradicting any potential folklore or superstitions about increased emergencies during lunar events.
